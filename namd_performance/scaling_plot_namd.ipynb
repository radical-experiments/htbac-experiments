```python
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
import pandas as pd

x = [8,16,32,64]
y1 = [6646.462402, 2685.136963, 1788.165039, 1033.362183]
y2 = [6641.104980, 2739.392090, 1790.020996, 990.429138]

for idx,i in enumerate(y1):
    new_i=i/float(60)
    y1[idx]=new_i

for idx,i in enumerate(y2):
    new_i=i/float(60)
    y2[idx]=new_i
    
y3 = np.array((y1,y2))
print y3
df = pd.DataFrame(columns=['Wallclock Time'])
df_err = pd.DataFrame(columns=['Wallclock Time'])
for i in range(len(x)):
    df.loc[str(x[i])] = np.mean([y1[i],y2[i]])
    df_err.loc[str(x[i])] = np.std([y1[i],y2[i]])/sqrt(2)

FONTSIZE=20
fig,axes = plt.subplots(nrows=1,ncols=1)
ax = df.plot(ax=axes, kind='bar', yerr=df_err, fontsize=FONTSIZE, legend=False)
fig.set_size_inches(20,10)
ax.set_title('NAMD Scaling using MPI', fontsize=FONTSIZE)
ax.set_xlabel('Number of Cores', fontsize=FONTSIZE)
ax.set_ylabel('WallClock Time (Mins)', fontsize=FONTSIZE)
plt.show()

print df_err

fig.savefig('plot_namd_scaling.pdf', dpi=100, bbox_inches='tight')
#fig.savefig('plot_namd_scaling.png', dpi=100, bbox_inches='tight')
```
